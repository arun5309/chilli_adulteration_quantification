```python
from PIL import Image
```

```python
import numpy as np
```

```python
def img_to_hist(imgo):
    imgo = imgo.resize((512, 512))
    imgo = np.asarray(imgo)
    imgo = imgo.transpose((2, 0, 1))
    imgs = [imgo[:, :256, :256], imgo[:, 256:, :256], imgo[:, :256, 256:], imgo[:, 256:, 256:]]
    return [np.array([np.histogram(k, bins=range(256), density=True)[0] for k in img]).ravel() for img in imgs]
```

```python
res = img_to_hist(Image.open('prepped/10/IMG_20211218_133435.jpg'))
```

```python
X, y = [], []

vals = [0] + list(range(50, 101, 5))

for i in vals:
    for p in os.listdir(f'prepped/{i}'):
        X += img_to_hist(Image.open(f'prepped/{i}/{p}'))
        y += [i] * 4
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(X)
```

```python
df.insert(len(df.columns), 'class', y)
```
